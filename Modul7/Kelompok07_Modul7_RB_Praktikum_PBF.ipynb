```python
"""
Langkah dalam menyelesaikan tugas ini adalah:
• Buatlah terlebih dahulu list dictionary yang kosong
• Buatlah looping untuk melakukan kalkulasi dengan fungsi Map atau Filter atau
Reduce kemudian simpan ke dalam list dictionary yang sudah di definisikan
Bagian 1: Model Populasi Polisi Detroit
Baca data dari file Laporan Polisi Detroit menggunakan modul csv dan terjemahkan
data ke dalam daftar dictionary. Menggunakan Filter dengan fungsi lambda untuk
mengecualikan dictionary (baris CSV) yang memiliki data yang hilang di kolom Zip,
atau kolom Neighborhood. Dengan menggunakan fungsi lambda dan Reduce, hitung
total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata
untuk kepolisian Detroit.
Bagian 2: Modelkan Neighborhood Samples
Menggunakan fungsi lambda dan map, atau lambda dan Filter, bagi list dictionary
menjadi list dictionary yang lebih kecil yang dipisahkan oleh neighborhood. Dengan
menggunakan lambda dan Reduce, temukan total waktu respons rata-rata untuk
setiap neighborhood, waktu pengiriman rata-rata untuk setiap neighborhood, dan total
waktu rata-rata untuk setiap neighborhood dan simpanlah ke dalam list dictionary.
Tambahkan item dictioanary untuk menyertakan data populasi untuk semua Detroit
dalam daftar gabungan Anda.

Bagian 3: Buat file Output JSON
Menggunakan modul JSON, format list dictionary Anda sebagai JSON
"""
```

```python
import pandas as pd
import json
from functools import reduce

#import dataset
df = pd.read_csv('./911_Calls_for_Service_(Last_30_Days).csv')

# cleaning data
df = df.dropna(subset=['Zip', 'Neighborhood'])

#mengubah dataset ke dalam bentuk list dictionary
data = df.to_dict('records')

data
```

```python
# BAGIAN 1 Model Populasi Polisi Detroit
# buat list dictionary kosong
list_dict_output = []
# Baca data dari file Laporan Polisi Detroit menggunakan modul csv dan terjemahkan
# data ke dalam list dictionary. Menggunakan Filter dengan fungsi lambda untuk
# mengecualikan dictionary (baris CSV) yang
# memiliki data yang hilang di kolom Zip,
# atau kolom Neighborhood. Dengan menggunakan fungsi lambda dan Reduce, hitung
# total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata
# untuk kepolisian Detroit.

# Menggunakan Filter dengan fungsi lambda untuk mengecualikan dictionary (baris CSV) yang memiliki data yang hilang di kolom Zip, atau kolom Neighborhood.
list_dict_output = list(
    filter(
        lambda x: x["zip_code"]
        or x["neighborhood"]
        or x["totalresponsetime"]
        or x["time_on_scene"]
        or x["totaltime"],
        data,
    )
)

# Dengan menggunakan fungsi lambda dan Reduce, hitung total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata untuk kepolisian Detroit.
# Menghitung total waktu respons rata-rata
total_response_time = reduce(
    lambda x, y: x + y,
    map(lambda x: x.get("totalresponsetime", 0), list_dict_output),
) / len(list_dict_output)

# # # Menghitung waktu pengiriman rata-rata
delivery_time = reduce(
    lambda x, y: x + y, map(lambda x: int(x.get("time_on_scene", 0)), list_dict_output)
) / len(list_dict_output)

# # # Menghitung total waktu rata-rata
total_time = reduce(
    lambda x, y: x + y, map(lambda x: int(x.get("totaltime", 0)), list_dict_output)
) / len(list_dict_output)

# # # print hasil dengan format keterangan
print("Total Response Time Average: ", total_response_time)
print("Delivery Time Average: ", delivery_time)
print("Total Time Average: ", total_time)
```

```python
# Bagian 2: Modelkan Neighborhood Samples
# Menggunakan fungsi lambda dan map, atau lambda dan Filter, bagi list dictionary
# menjadi list dictionary yang lebih kecil yang dipisahkan oleh neighborhood. Dengan
# menggunakan lambda dan Reduce, temukan total waktu respons rata-rata untuk
# setiap neighborhood, waktu pengiriman rata-rata untuk setiap neighborhood, dan total
# waktu rata-rata untuk setiap neighborhood dan simpanlah ke dalam list dictionary.
# Tambahkan item dictioanary untuk menyertakan data populasi untuk semua Detroit
# dalam daftar gabungan Anda.

# Menggunakan fungsi lambda dan map, atau lambda dan Filter, bagi list dictionary
# menjadi list dictionary yang lebih kecil yang dipisahkan oleh neighborhood.


# count unique neighborhood
neighborhoods = list(set(map(lambda x: x["neighborhood"], data)))


# group by neighborhood
neighborhood_dict = {}

for neighborhood in neighborhoods:
    neighborhood_dict[neighborhood] = list(
        filter(lambda x: x["neighborhood"] == neighborhood, data)
    )

# Dengan menggunakan lambda dan Reduce, temukan total waktu respons rata-rata untuk
# setiap neighborhood, waktu pengiriman rata-rata untuk setiap neighborhood, dan total
# waktu rata-rata untuk setiap neighborhood dan simpanlah ke dalam list dictionary.
for neighborhood in neighborhood_dict:
    neighborhood_dict[neighborhood] = {
        "total_response_time": reduce(
            lambda x, y: x + y,
            map(
                lambda x: x.get("totalresponsetime", 0), neighborhood_dict[neighborhood]
            ),
        )
        / len(neighborhood_dict[neighborhood]),
        "delivery_time": reduce(
            lambda x, y: x + y,
            map(
                lambda x: int(x.get("time_on_scene", 0)),
                neighborhood_dict[neighborhood],
            ),
        )
        / len(neighborhood_dict[neighborhood]),
        "total_time": reduce(
            lambda x, y: x + y,
            map(lambda x: int(x.get("totaltime", 0)), neighborhood_dict[neighborhood]),
        )
        / len(neighborhood_dict[neighborhood]),
    }


# Tambahkan item dictioanary untuk menyertakan data populasi untuk semua Detroit
# dalam daftar gabungan Anda.

neighborhood_dict["All Detroit"] = {
    "total_response_time": total_response_time,
    "delivery_time": delivery_time,
    "total_time": total_time,
}
```

```python
# BAGIAN 3

# Buat file Output JSON
# Menggunakan modul JSON, format list dictionary Anda sebagai JSON

with open("output.json", "w") as f:
    json.dump(neighborhood_dict, f)

with open("output.json", "r") as f:
    data = json.load(f)
# print hasil
print("Output JSON created")

```
